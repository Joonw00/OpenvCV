```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = np.full((512, 512, 3), 255, np.uint8)
image = cv2.line(image, (0, 100), (500, 255), (255, 0, 0), 1)

plt.imshow(image)
plt.show()
```

```python
#cv2.rectangle(image,start,end,color,thickness) : 하나의 사각형을 그리는 함수
#시작:왼쪽 위,  끝 : 오른쪽 아래 꼭짓점
image = np.full((512, 512, 3), 0, np.uint8)
image = cv2.rectangle(image, (20, 20), (255, 255), (255, 0, 0), -1)

plt.imshow(image)
plt.show()
```

```python
#cv2.circle(image,center,radidan,color,thickness) : 하나의 원을 그리는 함수
#radian : 반지름
image = np.full((1000, 1000, 3), 255, np.uint8)
image = cv2.circle(image, (500, 500), 250, (0, 0, 0), -1)

plt.imshow(image)
plt.show()
```

```python
#cv2.polylines(image,points,is_closed,color,thickness) : 하나의 다각형을 그리는 함수
#points : 꼭짓점들
#is_closed : 닫힌 도형 여부
image = np.full((512, 512, 3), 255, np.uint8)

#포인트에 대한 리스트 ->2차원 배열 형태
points = np.array([[0, 0], [400, 0], [0, 400], [500, 500], []])
image = cv2.polylines(image, [points], True, (0, 0, 255), 4)

plt.imshow(image)
plt.show()
```

```python
#cv2.putText(image,text,position,font_type,font_scale,color) : 하나의 텍스트를 그리는 함수
#position : 텍스트가 출력될 위치
image = np.full((512, 512, 3), 255, np.uint8)
image = cv2.putText(image, 'Hello World', (0, 200), cv2.FONT_ITALIC, 2, (255, 0, 0))

plt.imshow(image)
plt.show()
```
