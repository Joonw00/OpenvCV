cv2.findContours(image,mode,method) : 이미지에서 contour(외곽)들을 찾는 함수

mode : contour들을 찾는 방법

RETR_EXTERNAL : 바깥쪽 라인만 찾기

RETR_LIST : 모든 라인을 찾지만,Hierarchy 구성x

RETR_TREE : 모든 라인을 찾으며, 모든 hIERACHY 구성O

method : Contour들을 찾는 근사치 방법

CHAIN_APPROX_NONE : 모든 CONTOUR 포인트 저장

CHAIN_APPROX_SIMP : Contour Line을 그릴 수 있는 포인트만 저장

입력 이미지는 GRAY scale Threshold 전처리 과정이 필요함

```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('stair.jpg')
#흑백처리
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
#127이상255,이하0
ret, thresh = cv2.threshold(image_gray, 127, 255, 0)

plt.imshow(cv2.cvtColor(thresh, cv2.COLOR_GRAY2RGB))
plt.show()

#[1] 왜 안되지
contours = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)[0]
image = cv2.drawContours(image, contours, -1, (0, 0, 255), 4)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()

contours = cv2.findContours(thresh,cv2.RETR_LIST,cv2.CHAIN_APPROX_SIMPLE)[0]
image = cv2.drawContours(image, contours, 0, (255, 0, 0), 4)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()

contours = cv2.findContours(thresh,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)[0]
image = cv2.drawContours(image, contours, -1, (0, 255, 0), 4)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

cv2.drawContours(image,contours,contour_index,color,thickness) : Contour들을 그리는 함수

contour_index : 그리고자 하는 contourss line(전체 : -1)
